```julia
rmax = 150
rmin = 50
a_tick_angles  = collect(0:π/6:2π)[begin:end-1]
a_tick_texts   = 1:length(a_tick_angles)
a_tick_flag    = :out
r_tick_values  = range(rmin,rmax,5)
r_tick_texts   = range(0,100,5)
r_axis_angle = π/4


# Angle Axis
r_tick = a_tick_flag == :out ? rmax : rmin
out_circle = S(:fillOpacity=>0,:stroke=>:black)Circle(r=rmax)
in_circle = rmin > 0 ? S(:fillOpacity=>0,:stroke=>:black)Circle(r=rmin) : NilD()
a_ticks = mapreduce(t->R(t)T(r_tick,0)R(π/2)U(rmax/100)Tick(),+,a_tick_angles)

a_tick_texts = map(t->TextMark(text=t,anchor=:c,fontsize=7),a_tick_texts)
max_tick_texts_width = mapreduce(t->boundingwidth(t), max,a_tick_texts)

t_ticks = map(t-> begin
        r = r_tick + max_tick_texts_width
        if a_tick_flag == :in
            r = r_tick - max_tick_texts_width
        end
        x,y = r*cos(t),r*sin(t)
        T(x,y)
    end
        ,a_tick_angles)
a_tick_texts = reduce(+,t_ticks .* a_tick_texts)
a_ticks = a_ticks + a_tick_texts
a_grid  = mapreduce(a-> begin
        xinit,yinit = rmin * cos(a),rmin * sin(a)
        x,y = rmax*cos(a),r*sin(a)
        S(:stroke=>:grey,:strokeOpacity=>0.3)Line([[xinit,yinit],[x,y]])
    end,+,a_tick_angles
)

# Radius Axis
r_ticks= mapreduce(r->begin
            x,y = r[1]*cos(r_axis_angle), r[1]*sin(r_axis_angle)
            T(x,y)*(S(:stroke=>:white,:strokeWidth=>14)TextMark(text=r[2],fontsize=7)+TextMark(text=r[2],fontsize=7))
    end
        ,+,zip(r_tick_values, r_tick_texts))

r_grid = mapreduce(r->S(:fillOpacity=>0,:stroke=>:grey,:strokeOpacity=>0.5)Circle(r=r),+,r_tick_values)

d = out_circle + in_circle + a_grid + a_ticks + r_grid + r_ticks 
drawsvg(d, height=500)
```

```julia
PolarFrame(rmax=plt.figsize[1],)
# angle_scale.domain
function polarframe(plt)
    angle_scale = getscale(plt,:angle)
    r_scale = getscale(plt,:r)
    
    rmax = min(plt.figsize...)/2
    rmin= getnested(plt.config,[:guide,:rmin],0)
    
    a_tick_angles = angle_scale.codomain
    a_tick_texts = angle_scale.domain
    
    a_tick_flag=getnested(plt.config,[:guide,:a_tick_flag],:out)

    r_tick_values= getnested(plt.config,[:guide,:r_tick_values],
        range(r_scale.codomain[1], r_scale.codomain[2],
            getnested(plt.config,[:guide,:r_n_ticks], 5)
        )
    )
    
    r_tick_texts= getnested(plt.config,[:guide,:r_tick_texts],
        range(r_scale.domain[1], r_scale.domain[2],
            getnested(plt.config,[:guide,:r_n_ticks], 5)
        )
    )
    r_axis_angle= getnested(plt.config,[:guide,:r_axis_angle],π/2)
    
    return PolarFrame(rmax,
        rmin,
        a_tick_angles,
        a_tick_texts,
        a_tick_flag,
        r_tick_values,
        r_tick_texts,
        r_axis_angle,
    )
end

drawsvg(polarframe(plt))
```

```julia
struct PlotExpression
    f::Function # (PlotSpec,Data) -> TMark
end

function (pexpr::PlotExpression)(spec, data)
    return pexpr.f(spec, data)
end

function plotexpression(gexpr::GraphicExpression)
    return PlotExpression((spec, data) -> begin
        gexpr(scaledata(data))
    end)
end

plotexpression(scatter())
```

```julia
struct A{T}
   x::T 
end
struct Testing{T}
    x::T
    a::A{T}
    
    function Testing{T}(x::T) where T
        a = A{T}(x)
        return new(x, a)
    end
end


Testing(1,A(2))
Testing{Int}(1)
```

No Vega-Lite temos:
encoding := (channel, field, data-type, value, functions, scale, guide)

No Vizagrams temos:
encoding := (channel, field, data-type, scale)

O `data-type` eh usado somente para fazer inferencia do tipo de escala, logo, nao parece ser fundamental.

encodings = (
    :x = (field=:col_a, datatype=:n, scale= Linear()),
    :y = (field=:col_b, datatype=:q, scale= Linear()),
)


Guide{Coordinate} - Os estilo de guide varia de acordo com o tipo de coordenada.
O guide contem o outter frame e o inner frame com grid, eixos, e legendas.
O Guide contem o tamanho total do frame, tanto inner como outter.

As escalas dependem do tamanho do frame, OU, os frames dependeram das escalas.
Para desenhar o guide, precisamos de antemao saber as escalas, ja que elas determinam os eixos e legendas.
O problema aqui parece ser que, por exemplo, para definir uma escala precisamos do dominio (dado de entrada),
e do codominio, e este geralmente esta relacionado com o tamanho do frame do plot.
Podemos tornar a escala como independente, e fazer o guide derivado da escala. Para isso,
usamos o valor do codomino da escala para computar o "length" dos eixos.

Se pensarmos que o codominio de uma escala pode ser algo infinitio, como a reta, nao podemos
aferir o eixo pela escala. Ou seja, o codominio finito na verdade eh um "apetrecho" (ad-hoc) da escala,
e nao uma propriedade fundamental, da qual podemos derivar o eixo.

```julia
f(x) = sin(x)
g(x) = cos(√x)
x = 0:0.01:10
y = f.(x)
z = g.(x)

data=StructArray(x=x,b=y,z=z);
```

```julia
scales = Dict(
    :x=>Linear(domain=[-1,1],codomain=[1,1]),
    :r => Linear(domain=[-1,1],codomain=[0,10]),
    :angle => Linear(domain=[-1,1],codomain=[0,2π]),
)
encoding = Dict(:r=>:z,:angle=>:b)

polar(r,angle) = r .* [cos(angle), sin(angle)]
coord = Dict(:enc=>(:r,:angle),:map=>polar)

applyscales(scales, encoding, data, coord)
```

```julia
function context(plt,data,var, enc)
    context_scale = getscale(plt, enc)
    return context_scale(Tables.getcolumn(data,var))
end
function context(plt,data,var)
    enc = plt.encodings
    enc_var = get_encoding_variable(enc,var)
    context_scale = map(s->getscale(plt,s),enc_var)
    if length(context_scale) == 1
        return context_scale[1](Tables.getcolumn(data,var))
    end
    return map(s->s(Tables.getcolumn(data,var)),context_scale)
end

function (plt::Plot)(data,var)
    return context(plt,data,var)
end
function (plt::Plot)(data,var,enc)
    return context(plt,data,var,enc)
end
```

```julia
enc = plt1.encodings
var = :c
for k in keys(enc)
    if enc[k][:field] == var
        continue 
    end
end

function get_encoding_variable(encodings, field_var)
    # for k in keys(encodings)
    #     if encodings[k][:field] == field_var
    #         return k
    #     end
    # end
    ks = collect(keys(encodings))
    mapreduce(k->encodings[k][:field] == field_var ? k : [], vcat,ks)
end
```
