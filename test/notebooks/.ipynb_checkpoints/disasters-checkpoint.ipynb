```julia
using Pkg
Pkg.activate("../../.")
using TestEnv
TestEnv.activate()
using Revise
using Vizagrams
using DataFrames
```

```julia
using VegaDatasets
```

```julia
df = DataFrame(dataset("disasters"))
df = dropmissing(df);

df = filter(Cols(:Entity)=>x->x!="All natural disasters", df);
```

```julia
plt = Plot(
    title="Disasters",
    config = (xaxis=(grid=(flag=false,),),),
    figsize=(500,400),
    data=df,
    encodings=(
        x = (field = :Year, datatype = :q, guide=(lim=(1900,2017),)),
        y = (field = :Entity,datatype=:n, guide=(tickvalues=unique(sort(df.Entity, rev=true)),)),
        color = (field = :Entity,),
        size = (field = :Deaths,datatype=:q,scale = Linear(domain=(0,maximum(df.Deaths)),codomain=(2,30)))
    ),
    graphic = scatter(mark=S(:stroke=>:black,:opacity=>0.8)Circle())
)


drawsvg(plt, height=600)
```
