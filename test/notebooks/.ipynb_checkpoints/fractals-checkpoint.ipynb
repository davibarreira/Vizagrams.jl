```julia
function sierpinski(n)
    if n == 0
        return Polygon([(1,0), (0,0), (1/2, 1)])
    else
        t = sierpinski(n - 1)
        return S(:fill=>:blue)U(0.5)*(T(-1/2,0)t + T(0,1)t + T(1/2,0)t)
    end
end

drawsvg(mapreduce(n->sierpinski(n),→,0:6))
```
