```julia
using Pkg
Pkg.activate(".")
using Vizagrams
using DataFrames
using Random
using NetworkLayout
using Graphs
using StructArrays
using LinearAlgebra
```

```julia
Random.seed!(4)
g = barabasi_albert(60, 1)
x = map(x->x[1],spring(g))
y = map(x->x[2],spring(g))

nodes = StructArray(x=x,y=y,c=rand(length(x)))
edg = collect(edges(g));
```

```julia
plt = Plot(
    config=(guide=false,),
    data=nodes,
    encodings=(
        x=(field=:x,),
        y=(field=:y,),
        color=(field=:c,legend=(flag=false,)),
    ),
    graphic=∑() do row
        S(:fill=>row.color)T(row.x,row.y)Circle(r=5)
    end
)

graphic_edges = ∑() do row
    x_src,x_dst = getscale(plt,:x).([nodes[row.src][1],nodes[row.dst][1]])
    y_src,y_dst = getscale(plt,:y).([nodes[row.src][2],nodes[row.dst][2]])
    Line([[x_src,y_src],[x_dst,y_dst]])
end(edg)

plt = plt.spec + graphic_edges + graphic(plt) 
draw(plt)
```

```julia
plt = Plot(
    config=(guide=false,),
    data=nodes,
    encodings=(
        x=(field=:x,),
        y=(field=:y,),
        color=(field=:c,legend=(flag=false,)),
    ),
    graphic=∑() do row
        S(:fill=>row.color)T(row.x,row.y)Circle(r=3)
    end
)

graphic_edges = ∑() do row
    x_src,x_dst = getscale(plt,:x).([nodes[row.src][1],nodes[row.dst][1]])
    y_src,y_dst = getscale(plt,:y).([nodes[row.src][2],nodes[row.dst][2]])
    he = 2
    le = norm([x_dst,y_dst] - [x_src,y_src])
    α = 0
    pti = (1-α) * [x_src,y_src] + α*[x_dst,y_dst]
    
    α = 1 - (he+4)/le
    pte = (1-α) * [x_src,y_src] + α*[x_dst,y_dst]
    return Arrow(pts=[pti,pte],headsize=2)
end(edg)

plt = plt.spec + graphic_edges + plt.graphic(scaledata(plt)) 
draw(plt,height=400)
```

```julia
ngr = ∑() do row
    T(row[:x],row[:y])U(0.5)Circle(r=row.c)
end(nodes)

egr = ∑() do row
    x_src,y_src = nodes[row.src]
    x_dst,y_dst = nodes[row.dst]
    Line([[x_src,y_src],[x_dst,y_dst]])
end(edg)

gr = ngr + egr

draw(gr)
```
